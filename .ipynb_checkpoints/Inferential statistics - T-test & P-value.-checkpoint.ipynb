In a packing plant, a machine packs cartons with jars. It is supposed that a new machine will pack faster on the average than the machine currently used. To test that hypothesis, the times it takes each machine to pack ten cartons are recorded. The results, in seconds, are shown in the tables in the file files_for_lab/machine.txt. Assume that there is sufficient evidence to conduct the t test, does the data provide sufficient evidence to show if one machine is better than the other?

* lets create a DF of the numbers show of the txt file for each machine.

But first:

Null Hypothesis (H0): The population mean packing time of the New Machine is equal to the population mean packing time of the Old Machine

Alternatuve Hypothesis (H1): New Machine faster than Old Machine

```python
import pandas as pd
import math
```

```python
data = { 'New Machine': [42.1, 41, 41.3, 41.8, 42.4, 42.8, 43.2, 42.3, 41.8, 42.7], 'Old Machine':[42.7, 43.6, 43.8, 43.3, 42.5, 43.5, 43.1, 41.7, 44, 44.1]
}
```

```python
machinedf= pd.DataFrame(data)
```

```python
machinedf
```

Lets lets do the step by step.

Calculate the **mean**:

```python
mean_new_machine=machinedf['New Machine'].mean()
mean_old_machine=machinedf['Old Machine'].mean()

print('Mean New Macchine:', round(mean_new_machine,2) ),
print('Mean Old Machine:', round(mean_old_machine,2) )
```

Calculate the **Standart Deviation**:

```python
std_new_machine = machinedf['New Machine'].std()
std_old_machine = machinedf['Old Machine'].std()

print('Std New Macchine:', round(std_new_machine,4) ),
print('Std Old Machine:', round(std_old_machine,4) )
```

Lets calculate the **Variance** ( Square of Std):

```python
variance_new_machine = machinedf['New Machine'].var()
variance_old_machine = machinedf['Old Machine'].var()

print('Variance New Macchine:', round(variance_new_machine,4) ),
print('Variance Old Machine:', round(variance_old_machine,4) )
```

What is our **n**?

```python
n_new_machine = machinedf['New Machine'].count()
n_old_machine = machinedf['Old Machine'].count()

print('N of New Macchine:', round(n_new_machine,4) ),
print('N of Old Machine:', round(n_old_machine,4) )
```

Now we have all the data needed to calculate our **T-Value**:

```python
#lets start calculating the diferences between the means

t = (mean_new_machine - mean_old_machine) / math.sqrt((std_new_machine**2 / n_new_machine) + (std_old_machine**2 / n_old_machine))

print("Calculated t-value:",round(t,5))
```

So lets check our degree of freedom, with an Alpha of 0.05, and taking in account this is a one-tailed test (Our Null Hypothesis is (H0): Is the new machine faster than the old machine):
* n1+n2-k(number of n's)= degree of freedom.
So:
* 10+10-2=18

Lets check the T-Table:


Since the calculated t-value (-3.3972) is far more extreme (in the negative direction) than the critical t-value (1.734), we would reject the null hypothesis in favor of the alternative hypothesis. This suggests that there is a statistically significant difference between the "New Machine" and "Old Machine," and that the "New Machine" is likely to pack cartons faster, consistent with our hypothesis.

Lets go even further, lets calculate the **P-Value**:

```python
#lets name our Degrees of freedom 'df':

df= n_new_machine + n_old_machine -2 
print('Degrees of Freedom:',df)
```

```python
from scipy import stats
```

**P-Value**:

```python
p_value=stats.t.cdf(t,df=df)
print('P-Value:',round(p_value,6))
```

Null Hypothesis (H0): The population mean packing time of the New Machine is equal to the population mean packing time of the Old Machine 1 = 2.


Our significance level was of 0.05, and since  0.001606 is much much smaller than our alpha, we can reject the null hypothesis, and accept our alternative hypothesis (H1), that the New Machine is  faster than the  old Machine
